```python
import pandas as pd
import numpy as np
import os 
import folium
from folium import plugins
import rioxarray as rxr
import earthpy as et
import branca.colormap as cm
```

```python
#Import data
data = 'C:/Users/spit/OneDrive/Documenten/Jr Specialist UCSC/Adina/Trace Metal Project/ES_sample_sites.csv'
sites = pd.read_csv(data)
sites.columns = ['Name', 'ID', 'Area', 'Lat', 'Lon', 'Sal', 'FTF', 'FTF2']

#Make list of coordinates
locations = sites[['Lat', 'Lon']]
locationlist = locations.values.tolist()
```

```python
#create a color map
color_var = 'Sal'
cmap = cm.LinearColormap(['blue', 'red'],
                         vmin=sites[color_var].quantile(0.05), vmax=sites[color_var].quantile(0.95),
                         caption = color_var)

#make map
m = folium.Map(location=[36.800052, -121.781338], zoom_start=11.5, tiles = 'Stamen Terrain', width=400, height=600)
m.add_child(cmap)#Add the color map legend to your map
def plotDot(point):
    folium.CircleMarker(location=[point.Lat, point.Lon], 
                        color=cmap(point[color_var]),
                        radius=2, 
                        weight=4).add_to(m)
sites.apply(plotDot, axis = 1)
m


```
